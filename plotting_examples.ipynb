### Plot the graph of

$y = x^2 + \frac{1}{x}, \quad \text{for } -3 \leq x \leq 3, \quad x \neq 0$

with intervals of 0.5

```python
x1 = np.arange(-3, 0, 0.5)
x2 = np.arange(0.5, 3.5, 0.5)
```

```python
y1 = x1**2 + 1/x1
y2 = x2**2 + 1/x2
```

```python
# Smooth curves
x1_s = np.linspace(-3, -0.5, num=50, endpoint=True)
x2_s = np.linspace(0.5, 3, num=50, endpoint=True)
y1_s = interp1d(x1, y1, kind='cubic')
y2_s = interp1d(x2, y2, kind='cubic')
```

```python
plt.plot(x1_s, y1_s(x1_s), color='b')
plt.plot(x2_s, y2_s(x2_s), color='b')
center_spines()
```

### Plot the graph of

$y = \begin{cases} x^2 + x + 1 & \text{for } -3 \le x \le 1 \\ 3 - x & \text{for } 1 < x \le 3 \end{cases}$

with intervals of 0.5

```python
x1 = np.linspace(-3, 1, num=20, endpoint=True)
x2 = np.linspace(1.01, 3, num=20, endpoint=True)
y1 = x1 ** 2 + x1 + 1
y2 = 3 - x2
```

```python
# Smooth curves
y1_s = interp1d(x1, y1, kind='cubic')
y2_s = interp1d(x2, y2, kind='cubic')
```

```python
plt.plot(x1, y1_s(x1), color='b')
plt.plot(x2, y2_s(x2), color='b')
center_spines()
```

### Plot the graph of

$y = \pm \sqrt{1 - x^2} \text{ for } -1 \le x \le 1$

with intervals of 0.5.

Use `pandas` to connect negative and positive parts of the graph and plot.

```python
x = np.arange(-1, 1.2, 0.2)
```

```python
y1 = np.sqrt(1-x**2)
y2 = np.negative(y1)
```

```python
graph_pos = np.vstack((x, y1)).T
graph_neg = np.vstack((x, y2)).T
graph = np.concatenate((graph_pos, graph_neg))
```

```python
df = pd.DataFrame(data=graph).round(decimals=3)
```

```python
df
```

```python
df.plot(x=0, y=1)
center_spines()
```

### Plot the graph of

```python
x1 = np.arange(-4, 0, 0.3)
x2 = np.arange(0.3, 4.3, 0.3)
y1 = 2-x1
y2 = np.power(x2, 2) + 2
```
